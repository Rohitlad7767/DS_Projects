```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\hp\Downloads\Crop_recommendation.csv")
```

```python
dataset
```

```python
dataset["label"].unique()
```

```python
dataset["label"].replace({'rice':'1', 'maize':'2', 'chickpea':'3','kidneybeans':'4','pigeonpeas':'5','mothbeans':'6','mungbean':'7','blackgram':'8','lentil':'9','watermelon':'10','muskmelon':'11','cotton':'12','jute':'13'},inplace=True)
```

```python
dataset
```

```python
dataset["season"].replace({'rainy':'1', 'winter':'2', 'spring':'3','summer':'4'},inplace=True)
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(dataset[['temperature','humidity','ph','water availability','season']],dataset['label'],test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)   #For checking score
```
